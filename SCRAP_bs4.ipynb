```python
import time
import random
import requests
import pandas as pd
from bs4 import BeautifulSoup as bsoup
```

```python
#Connection with the website
url = 'https://www.tripadvisor.com/Hotels-g54273-oa30-Hilton_Head_South_Carolina-Hotels.html'
#Read and connect with raw data URL
html_ua1 = 'https://gist.githubusercontent.com/pzb/b4b6f57144aea7827ae4/raw/cf847b76a142955b1410c8bcef3aabe221a63db1/user-agents.txt'
html_ua = requests.get(html_ua1)

'''
#In case you want to create an specific list of user agents. Check the url in "h
user_agents =['Mozilla/5.0 (Windows NT 6.3; WOW64; rv:40.0) Gecko/20100101 Firefox/40.0',
              'Mozilla/5.0 (Windows NT 6.1; WOW64; Trident/7.0; rv:11.0) like Gecko',
              'Mozilla/5.0 (iPad; CPU OS 8_3 like Mac OS X) AppleWebKit/600.1.4 (KHTML, like Gecko) Version/8.0 Mobile/12F69 Safari/600.1.4'
              ]
'''

#Creates a list of user agents contained in the raw url of "html_ua1"
user_agents = html_ua.text.splitlines()

#Randomly select a user agent from the list, selecting only 30 values of the total.
headers = {
            'User-Agent': random.choice(user_agents)
        }

#Stablish connection with "url" variable
html = requests.get(url, headers = headers)
time.sleep(5)
html.status_code #200 value is good, 204 is good but there is no content,3xx is redirection state, 4xx is client error, 5xx is server error.
```

```python
#Run the parser to read html page code
soup = bsoup(html.content,'lxml') #could be also "html.parser"
```

```python
#Search for specific content and storage it in a list
restaurant_names = []
n = 100
for name in soup.find_all('a', {'class': 'BMQDV _F Gv wSSLS SwZTJ FGwzt ukgoS'}, limit = n): #in 'div',{'class':'listing_title'} you will have to add the path in the html code depending on the content you need to extract
  restaurant_names.append(name.text.strip())
```

```python
hotel_costs = []
for name in soup.find_all('span', class_= 'YECgr'):
  hotel_costs.append(name.text.strip())
```

```python
reviews = []
for review in soup.find_all('span',{'class':'tvdtW'}):
  reviews.append(review.text.strip())
```

```python
ratings = []
for rating in soup.find_all('a',{'class':'luFhX o W f u w JSdbl'}):
  ratings.append(rating['alt'])
```
